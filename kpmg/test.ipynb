```python
import pandas as pd
```

```python
# Importing data
xls = pd.ExcelFile('KPMG_VI_New_raw_data_update_final.xlsx')

Transactions = pd.read_excel(xls, 'Transactions', skiprows=1)
NewCustomerList = pd.read_excel(xls, 'NewCustomerList', skiprows=1)
Demographic = pd.read_excel(xls, 'CustomerDemographic', skiprows=1)
Address = pd.read_excel(xls, 'CustomerAddress', skiprows=1)
```

```python
# Assuming Transactions is a DataFrame containing the 'product_first_sold_date' column

Transactions['product_first_sold_date'] = pd.to_timedelta(Transactions['product_first_sold_date'], unit='D') + pd.Timestamp('1900-01-01')
today_date = pd.Timestamp('2018-01-01')  # Set today's date as 2018-01-01
# Calculate the difference between today's date and the reference date
date_difference = pd.Timestamp.today() - today_date

# Subtract the date difference from the product_first_sold_date
Transactions['product_first_sold_date'] = Transactions['product_first_sold_date'] - date_difference

```

```python
# Make a product age column
Transactions["product_age"] = ((today_date - Transactions["product_first_sold_date"]).dt.days // 365.25)
```
